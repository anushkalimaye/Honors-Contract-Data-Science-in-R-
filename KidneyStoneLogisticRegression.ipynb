```r
#importing file
data <- read.csv("kidney_stone_urine_analysis.csv")
data
#file details
# 80 Samples
# Parameters: (8)
# Specific gravity: Density of urine relative to water
# pH
# Osmolarity: solute particles per unit of volume
# Conductivity: conductivity of urine
# Urea : concentration of urea in urine→micromoles/liter
# nitrogen-containing waste substance that the kidneys clear from the blood and excrete into the urine
# Calcium Concentration: micromoles/liter
# Target: 0- absence of stone 1- presence of stone

summary(data)
```

```r
#histogram visualization
parameters<-list(data$gravity, data$ph, data$osmo, data$cond, data$urea,data$calc,data$target)
colors <- c("red", "green", "blue", "orange", "purple", 'pink', 'yellow')
titles<- c('Specific Gravity', 'pH','Osmolarity (Osmols/L)', 'Conductivity (millimho) ', 'Urea (mmol/Liter)', 'Calcium (millimoles/Liter)', 'Target (0 or 1)')
n<-length(parameters)
histograms<-list()
for (i in 1:n){
  histogram <- hist(parameters[[i]], freq= FALSE, col=colors, xlab= titles[[i]], main= paste('distrubution of', titles[[i]]))
  histograms[[i]] <- histogram
  mean_value <- mean(parameters[[i]])
  abline(v = mean_value, col = 'red', lwd = 2)
  legend('topright', legend = paste('Mean:', round(mean_value, 2)), col = 'red', lwd = 2)
}



```

```r
#splitting into training and testing data


test_indices <- sample(nrow(data), size = 0.2 * nrow(data))  # 20% for testing

training <- data[-test_indices, ]
training
testing <- data[test_indices, ]
testing
```

```r
logistic_1t <- glm(target~ph+urea+gravity, data=training, family= 'binomial')

predicted_probabilities <- predict(logistic_1t, newdata = testing, type = "response")

predicted_classes <- ifelse(predicted_probabilities > 0.56, 1, 0)
actual_classes <- testing$target

predicted_classes <- factor(predicted_classes)
actual_classes <- factor(actual_classes)


conf_matrix <- confusionMatrix(predicted_classes, actual_classes)
print(conf_matrix)

roc_obj <- roc(testing$target, predicted_probabilities)

plot(roc_obj, main = paste("ROC Curve for", 'pH, urea, gravity'))
auc_value <- auc(roc_obj)

cat("AUC Value:", auc_value, "\n")
```
